```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("/content/Shopping_data.csv")
df
```

```python
'''
1---Unsupervised Data


2--- Category Column - "Genre"
 --- Continuous column - "CustomerID", "Annual Income (k$)", "Spending Score (1-100)","Age"

3---Correlation Test for Numerical columns

4--- Hypothesis testing: continuous Vs continuous
      customerid-annual income
      customerid-spendingscore
      customerid-age
      annualincome-spendingscore
      annualincome-age
      spendingscore-age
5---Hypothesis testing for category VS continous
      genre vs customerid
      genre vs annual income
      genre vs spending score
      genre vs age
```

```python
'''
Correlation Test for Numerical Columns
By the result of correlation Test,
No related columns -    customer id with age,
                        customer id with spending score,
                        annual income with age,
                        spending score with annual income

Related columns - costomer id and annual income,
                    age and spending score'''
```

```python
#Hypothesis Testing

#1-Two sample ttest for CustomerID VS spendinscore

for i in range(25):
  sample1 = df.CustomerID.sample(frac=0.05)
  sample2 = df["Spending Score (1-100)"].sample(frac=0.04)

  t_test, p_value = stats.ttest_ind(sample1,sample2)
  if p_value > 0.05:
    print("H0 null hypothesis accepted and Ha alternative hypothesis rejected")
  else:
    print("H0 null hypothesis rejected and Ha alternative hypothesis accepted")

# Here I got 13 acceptance and 12 rejection so that Null hypothesis is accepted by the majaricy results
```

```python
#Hypothesis Testing

#1-Two sample ttest for CustomerID VS age

for i in range(25):
  sample1 = df.CustomerID.sample(frac=0.05)
  sample2 = df["Age"].sample(frac=0.04)

  t_test, p_value = stats.ttest_ind(sample1,sample2)
  if p_value > 0.05:
    print("H0 null hypothesis accepted and Ha alternative hypothesis rejected")
  else:
    print("H0 null hypothesis rejected and Ha alternative hypothesis accepted")

# Here I got 7 acceptance and 15 rejection so that Null hypothesis is rejected by the majaricy results
```

```python
#Hypothesis Testing

#1-Two sample ttest for CustomerID VS annual income

for i in range(25):
  sample1 = df.CustomerID.sample(frac=0.05)
  sample2 = df["Annual Income (k$)"].sample(frac=0.04)

  t_test, p_value = stats.ttest_ind(sample1,sample2)
  if p_value > 0.05:
    print("H0 null hypothesis accepted and Ha alternative hypothesis rejected")
  else:
    print("H0 null hypothesis rejected and Ha alternative hypothesis accepted")

# Here I got 13 acceptance and 12 rejection so that Null hypothesis is accepted by the majaricy results
```

```python
# f-Test for continuous vs category

group = df['Genre'].unique()
data = {}
for i in group:
  data[i]=df["Annual Income (k$)"][df["Genre"]==i]


f_value, p_value = stats.f_oneway(data['Male'],data['Female']
                                  )
print(f_value, p_value)
if p_value<0.05:
  print("H0 - Null hypothesis is rejected which means no relationship exists")
else:
  print("Ha - Null hypothesis is accepted which means relationship exists")
#Result --Ha - Null hypothesis is accepted which means relationship exists
```

```python
# f-Test for continuous vs category

group = df['Genre'].unique()
data = {}
for i in group:
  data[i]=df["Age"][df["Genre"]==i]


f_value, p_value = stats.f_oneway(data['Male'],data['Female'])

print(f_value, p_value)
if p_value<0.05:
  print("H0 - Null hypothesis is rejected which means no relationship exists")
else:
  print("Ha - Null hypothesis is accepted which means relationship exists")
#Result --Ha - Null hypothesis is accepted which means relationship exists
```
