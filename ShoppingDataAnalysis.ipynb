```python
#Hypothesis Testing

#1-Two sample ttest for CustomerID VS spendinscore

for i in range(40):
  sample1 = df.CustomerID.sample(frac=0.05)
  sample2 = df["Spending Score (1-100)"].sample(frac=0.04)

  t_test, p_value = stats.ttest_ind(sample1,sample2)
  if p_value > 0.05:
    print("H0 null hypothesis accepted and Ha alternative hypothesis rejected")
  else:
    print("H0 null hypothesis rejected and Ha alternative hypothesis accepted")

# Here I got 17 acceptance and 23 rejection so that Null hypothesis is rejected by the majaricy results
```

```python
# f-Test for continuous vs category

group = df['Genre'].unique()
data = {}
for i in group:
  data[i]=df["Annual Income (k$)"][df["Genre"]==i]


f_value, p_value = stats.f_oneway(data['Male'],data['Female']
                                  )
print(f_value, p_value)
if p_value<0.05:
  print("H0 - Null hypothesis is rejected which means no relationship exists")
else:
  print("Ha - Null hypothesis is accepted which means relationship exists")
#Result --Ha - Null hypothesis is accepted which means relationship exists
```

```python
# f-Test for continuous vs category

group = df['Genre'].unique()
data = {}
for i in group:
  data[i]=df["Age"][df["Genre"]==i]


f_value, p_value = stats.f_oneway(data['Male'],data['Female'])

print(f_value, p_value)
if p_value<0.05:
  print("H0 - Null hypothesis is rejected which means no relationship exists")
else:
  print("Ha - Null hypothesis is accepted which means relationship exists")
#Result --Ha - Null hypothesis is accepted which means relationship exists
```
